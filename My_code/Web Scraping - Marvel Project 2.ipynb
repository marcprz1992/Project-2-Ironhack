```python
def getLinks (url):
    html_1 = requests.get(url)
    soup_1 = BeautifulSoup(html_1.content, "html.parser")
    tags_1 = soup_1.find_all("a") # find_all targets divs
    links_1 = [tag.get("href") for tag in tags_1]
    return [i for i in links_1[1:] if "http" in i]
```

## As it works, let's create a comprehension list so all the wiki pages can be included

```python
list_of_links = [i for i in df_and_api_final["char"] ]
list_of_links
```

## After manually trying to run it for some characters, I realised the "wikidata" page needs to be amended in order for it to perform properly

```python
list_of_links_two = [i.replace("http","https").replace("entity","wiki") for i in list_of_links]
list_of_links_two    
```

## Now we can define an empty list and run a nested for loop with a conditional function in order to get the gender of each superhero via web scrapping

```python
gender_list = []

for url in list_of_links_two:
    html = requests.get(url)
    soup = BeautifulSoup(html.content, "html.parser")
    products = soup.find_all("div", attrs={"class":"wikibase-snakview-value wikibase-snakview-variation-valuesnak"})
    for i in range(len(products)):
        if products[i].getText() == "male":
            gender_list.append(products[i].getText())
        elif products[i].getText() == "female":
            gender_list.append(products[i].getText())
        else:
            pass
gender_list
```

```python
len(gender_list)
```

## Let's create a new dictionary with the gender variable that will then be concatenated with the main dataframe

```python
web_scrap_col = {
    "gender": gender_list
}
```

```python
web_scrap = pd.DataFrame(web_scrap_col)
web_scrap
```

```python
ultra_merged_df = pd.concat([df_and_api_final, web_scrap], axis=1)
ultra_merged_df
```
